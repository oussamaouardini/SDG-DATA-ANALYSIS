## Introduction

The emergence of Corona Virus disease (COVID-19) has led the world to an unprecedented public health crisis, which affected the lives of all sections of society as people were asked to self-quarantine in their homes to prevent the spread of the virus. The lockdown had serious implications on mental health, resulting in psychological problems including frustration, stress, and depression. 

The students from the less privileged backgrounds have experienced larger negative impacts due to the Covid-19 outbreak such as Reduction in family income, limited access to digital resources, and the high cost of internet connectivity have disrupted the academic life of the students. Moreover, changes in daily routine including lack of outdoor activity, disturbed sleeping patterns, social distancing have affected the mental well-being of the students.

In this project, we investigated and analyzed the potential consequences of the COVID-19 pandemic on the life of students. Moreover, our study attempts to assess the mental situation of students of different age groups using different parameters including sleeping habits, daily fitness routine, and social support. 


At the first point we started looking for the dataset especially who is realated to the students life with COVID-19, as we found it in this kaggle https://www.kaggle.com/kunal28chaturvedi/covid19-and-its-impact-on-students 

This dataset represents COVID-19'impact on education, social life and mental health of students & also the students' behaviour when they are in the lockdown 2020.

# Data Pre-Processing

```python
# Replasing values in "Time spent on TV" data
df['Time spent on TV'].replace('No tv', 0, inplace=True)
df['Time spent on TV'].replace('n', 0, inplace=True)
df['Time spent on TV'].replace('N', 0, inplace=True)
df['Time spent on TV'].replace(' ', 0, inplace=True)
df['Time spent on TV'] = df['Time spent on TV'].astype('float')
df['Prefered social media platform'] = df['Prefered social media platform'].replace('None ', 'None')
df['Prefered social media platform'] = df['Prefered social media platform'].replace('Whatsapp', 'WhatsApp')

# function to help in replasing miss written data in the column of "What you miss the most"
def missing_col(value):
    if 'nothing' in value.lower():
        return 'Nothing'
    elif 'all' in value.lower() or 'everything' in value.lower():
        return 'All'
    elif 'friends' in value.lower() or 'friend' in value.lower():
        return 'Friends'
    else:
        return value
    
```

for simplification reason we will change the columns name in the list rename

```python
rename = ['ID','region', 'age', 'class_time', 'rating', 'class_medium', 'self_study_time', 'fitness_time', 'sleep_time', 'social_time', 'social_media', 'tv_time', 'meals_per_day', 'weight_change', 
          'health_issue', 'stress_busters', 'utilized_time', 'is_social', 'missing']
```

```python
original = dict(zip(df.columns, rename))
```

```python
df.columns = rename
df.head()
```

```python
# replace miss written values in the column of missing by applying the function missing_col
df.missing = df.missing.apply(missing_col)
```

```python
# Calualate the sum of features that containes null values
df.isna().any().sum()
```

the raiting column & class_medium who containes the missing values

```python
# A visualisation using Seaborn to check missing values 
sns.heatmap(df.isna())
```

```python
# to erease null values, we define this function impute_na to fill the missing values with most frequent one in each category
def impute_na(data, variable):
    most_frequent_category = data.groupby([variable])[variable].count().sort_values(ascending=False).index[0]
    data[variable].fillna(most_frequent_category, inplace=True)
```

```python
for variable in ['rating', 'class_medium']:
    impute_na(df, variable)
```

```python
# verify if still there is missing values
df.isna().any()
```

```python
sns.heatmap(df.isna())
```

### Students Weights Increase

What is causing increase in the weight ?

To interpret this, we can say that - The relation between the meals they take and hours they sit (class time + social time + self study + sleep) with no action

```python
# we add a new column colled sitting_hours that counts number of sitting hours ( no action ) to see if it may be responsible for students weights
sitting_hours = df.sleep_time + df.social_time + df.class_time + df.self_study_time
temp1 = pd.concat([df, sitting_hours.rename('sitting_hours')], axis= 1)
temp1.sample()
```

```python
# to adjust the values in sitting_hours column
temp1.sitting_hours = temp1.sitting_hours.apply(lambda x: 24 if x > 24 else x)
temp1.sitting_hours = temp1.sitting_hours.apply(round)
```

```python
temp1.head()
```

```python
temp1.sitting_hours.value_counts()
```

```python
sns.countplot(x= 'sitting_hours', data= temp1)
plt.title("Sitting Hour with SLEEP time")
```

we will show an other plot and recalculate the sitting_hours without sleep time

```python
sitting_hours = df.social_time + df.class_time + df.self_study_time
temp2 = pd.concat([df, sitting_hours.rename('sitting_hours')], axis= 1)

temp2.sitting_hours = temp2.sitting_hours.apply(lambda x: 24 if x > 24 else x)
temp2.sitting_hours = temp2.sitting_hours.apply(round)
sns.countplot(x= 'sitting_hours', data= temp2)
plt.title("Sitting Hour without SLEEP time");
```

<b>Graph changes</b> So, let's not consider sleep time for now.

Next up, we will see - that if weight increase, decrease is related with more sitting 

```python
sns.kdeplot(x= "sitting_hours", data= temp2, hue= 'weight_change')
plt.xticks(range(1, 25));
```

Students who are sitting morethan 6 hours are tend to have the weight increase, but there we can't see a clear cut because more than those who have reported that their weight has increased if they sit more, there are those whose weight has remain the same and even sitting more than 6 hours. So here, also is not a clear explanation of relation between sitting and weight increase, decrese.

```python
(pd.crosstab(temp2.sitting_hours, temp2.weight_change, normalize= 'columns') * 100).round(1)
```
